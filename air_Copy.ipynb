```python
median_gas_PM25=f['PM2.5'].median()
median_gas_PM10=f['PM10'].median()
median_gas_NO=f['NO'].median()
median_gas_NO2=f['NO2'].median()
median_gas_NOx=f['NOx'].median()
median_gas_NH3=f['NH3'].median()
median_gas_CO=f['CO'].median()
median_gas_SO2 =f['SO2'].median()
median_gas_O3=f['O3'].median()
median_gas_Benzene=f['Benzene'].median()
median_gas_Toluene =f['Toluene'].median()
median_gas_Xylene=f['Xylene'].median()
median_gas_AQI =f['AQI'].median()
```

```python
f['PM2.5']=f['PM2.5'].fillna(median_gas_PM25)
f['PM10']=f['PM10'].fillna(median_gas_PM10)
f['NO']=f['NO'].fillna(median_gas_NO)
f['NO2']=f['NO2'].fillna(median_gas_NO2)
f['NOx']=f['NOx'].fillna(median_gas_NOx)
f['NH3']=f['NH3'].fillna(median_gas_NH3)
f['CO']=f['CO'].fillna(median_gas_CO)
f['SO2']=f['SO2'].fillna(median_gas_SO2)
f['O3']=f['O3'].fillna(median_gas_O3)
f['Benzene']=f['Benzene'].fillna(median_gas_Benzene)
f['Toluene']=f['Toluene'].fillna(median_gas_Toluene)
f['Xylene']=f['Xylene'].fillna(median_gas_Xylene)
f['AQI']=f['AQI'].fillna(median_gas_AQI)
f['AQI_Bucket'] = f.groupby('City')['AQI_Bucket'].ffill().bfill()
```

```python
f.isnull().sum()
```

```python
f.to_csv('mydata.csv')
```

```python
import matplotlib.pyplot as plt

gases = ['NO2', 'PM2.5', 'PM10', 'SO2', 'CO', 'O3']

avg_gases = a.groupby('City')[gases].median()

avg_gases.plot(kind='bar', figsize=(12,6))

plt.title("Average Gases in Each City")
plt.xlabel("City")
plt.ylabel("Average Concentration")
plt.xticks(rotation=45)
plt.legend(title="Gases")

```

```python
top4 = a.groupby("City")["AQI"].mean().nlargest(5)

colors = ['#ff4d4d', '#ff944d', '#ffd11a','#008000','#66cc66']

top4.plot(kind='barh', figsize=(8,5),color=colors)

plt.xlabel("Average AQI")
plt.title("Top 5 Polluted Cities")
plt.gca().invert_yaxis()
```

```python
a['AQI_Bucket'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(5,5))
plt.title("AQI Bucket Distribution")
plt.ylabel("")
```

```python
a['Year'] = pd.to_datetime(a['Date']).dt.year

trend = a.groupby("Year")["AQI"].mean()

trend.plot(marker='o', figsize=(10,5))

plt.title("AQI Trend Overview (2015–2020)")
plt.xlabel("Year")
plt.ylabel("Average AQI")
plt.grid(True)
```
